# Text to Image



diffusers, bir Python kütüphanesidir ve resim sentezi için kullanılır. Bu kütüphane, OpenAI'nin DALL-E modelinin temelindeki teknolojiyi kullanarak, belirli bir metin prompt'u kullanarak resimlerin otomatik olarak oluşturulmasını sağlar.

```python
from diffusers import StableDiffusionPipeline    # diffusers kütüphanesinden StableDiffusionPipeline sınıfını içe aktarır
import torch                                    # torch kütüphanesini içe aktarır

model_id = "runwayml/stable-diffusion-v1-5"      # kullanılacak modelin kimliğini belirler
pipe = StableDiffusionPipeline.from_pretrained(model_id, torch_dtype=torch.float16)   # belirtilen modeli yükleyerek bir StableDiffusionPipeline nesnesi oluşturur
pipe = pipe.to("cuda")                          # oluşturulan nesneyi CUDA cihazına taşır (GPU kullanımı için)
```

```python
# Kullanıcıdan prompt'u al
prompt = input('Resim oluşturmak için prompt giriniz: Çıkmak için q tuşuna basınız: ')

# Eğer prompt boş ise varsayılan bir prompt belirle
if prompt == '':
    prompt = "a photo of an astronaut riding a horse on mars"

# Prompt'u kullanarak resim sentezi yap
image = pipe(prompt).images[0]

# Oluşturulan resmi "image.png" olarak kaydet
image.save("image.png")

from IPython.display import Image
# Görüntüyü dosya yolundan yükle ve göster
image_path = "/content/image.png"
Image(filename=image_path)
```
