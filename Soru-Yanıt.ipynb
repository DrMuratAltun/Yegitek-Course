# Yapay Zekâ Sohbet Asistanı
Bu uygulama Hugging Face üzerinden (Büyük Dil Modeli) LLM kullanarak kıendi sohbet botumuzu yapacağız.

Python Transformers kütüphanesi, Hugging Face tarafından geliştirilmiştir ve açık kaynaklıdır. Kütüphane, önceden eğitilmiş dil modellerini indirme, yükleme ve kullanma işlemlerini basitleştirir. Ayrıca, bu modelleri özelleştirmek, ince ayar yapmak ve transfer öğrenme uygulamaları için kullanmak için kullanışlı araçlar sağlar.

"pipeline" modülünü içe aktararak, birkaç satır kod ile önceden eğitilmiş bir model pipeline örneği oluşturabilirsiniz. Bu pipeline, çok miktarda ön işlem kodu yazmadan belirli NLP görevlerini gerçekleştirmek için kullanılabilir.

Bu kod parçacığı, "pipeline" modülünü kullanarak bir sohbet robotu oluşturur. Sohbet robotu, "conversational" görevi için önceden eğitilmiş bir model olan "facebook/blenderbot-400M-distill" modelini kullanır. Bu model, Facebook tarafından geliştirilen ve geniş bir metin korpusunda eğitilmiş olan BlenderBot adlı sohbet robotunun 400 milyon parametreli ve küçültülmüş bir sürümüdür.

```python
# transformers kütüphanesini kur
!pip install transformers
```

```python
from transformers import AutoTokenizer, AutoModelForQuestionAnswering, pipeline

tokenizer = AutoTokenizer.from_pretrained("savasy/bert-base-turkish-squad")
model = AutoModelForQuestionAnswering.from_pretrained("savasy/bert-base-turkish-squad")
chatbot = pipeline("question-answering", model=model, tokenizer=tokenizer)
```
